# Linearn Regression

## Importing Libraries

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

## Reading Data for android

```python
android_data = pd.read_csv('android_data_modified.csv')
android_data.head()
```

```python
linear_model_android = LinearRegression()
linear_model_android
```

```python
android_data.iloc[:, [0,2,3,4,5,6,7,8]]
```

```python
android_data.iloc[:,1]
```

```python
linear_model_android.fit(android_data.iloc[:, [2,3,4,5,6,7,8]], android_data.iloc[:,1])
```

```python
android_data.iloc[-1,2:]
```

```python
linear_model_android.predict(android_data.iloc[-1,2:])
```
